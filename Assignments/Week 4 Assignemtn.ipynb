```python
import csv
import matplotlib.pyplot as plt


class FOLK1A():
    def __init__(self, location, year, amount):
        self.location = location
        self.year = year
        self.amount = amount

    def __repr__(self):
        return "{" + str(self.location) + ", " + str(self.year) + ", " + str(self.amount) +  "}"



def read_FOLK1A_from_csv():
    with open('./FOLK1A.csv', newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=',' , quotechar='"')
        for row in reader:            
            print(', '.join(row))



def get_FOLK1A_from_csv():
    with open('./FOLK1A.csv', newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=',' , quotechar='"')
        folk1Alist = []
        flag = True

        for row in reader:
            if(flag):
                flag = False
            else:    
                data = (','.join(row))
                datalist = data.split(";")      
                folk1a = FOLK1A(datalist[0], datalist[2], float(datalist[3]))
                folk1Alist.append(folk1a)
        return folk1Alist



def plot_FOLK1A(folk1aList):
    time = []
    amount = []
    for folk1a in folk1aList:
        folk1ayear = folk1a.year
        year = ""
        yearfloat = 0

        for i in range(0,4):
            year += (folk1ayear[i])
        
        yearfloat = float(year)

        if (folk1ayear[5] == "2"):
            yearfloat += 0.25
        if (folk1ayear[5] == "3"):
            yearfloat += 0.5
        if (folk1ayear[5] == "4"):
            yearfloat += 0.75

        amount.append(folk1a.amount)
        time.append(yearfloat)
    plt.plot(time, amount)
    plt.xlabel('Years')
    plt.ylabel('Amount of divorced people')





#read_FOLK1A_from_csv()
#print(get_FOLK1A_from_csv())
plot_FOLK1A(get_FOLK1A_from_csv())
```
